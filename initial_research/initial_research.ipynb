```python
import requests
import json
import pandas as pd
from config import c_key
#run pip install nltk in your terminal with your PythonData environment activated if you have not already installed nltk
import nltk 
#nltk.download('vader_lexicon') <- you will need to run this the first time you run this code 
from nltk.sentiment.vader import SentimentIntensityAnalyzer
SID = SentimentIntensityAnalyzer()
```

## Currents API Testing

Documentation: https://currentsapi.services/en

Attempted to search by keywords "immigrants", "refugees", and "migrants" but found that results were the same for all three searches.

```python
#Keyword search for "immigrants"
base_url = 'https://api.currentsapi.services/v1/search?language=en'
country_1 = 'us'
keyword_1 = 'immigrants'
url_1 = (f"{base_url}&country={country_1}&keywords={keyword_1}&apiKey={c_key}")
response_1 = requests.get(url_1)
data_1 = response_1.json()
print(json.dumps(data_1, indent=4))
```

```python
#Store data in list of dictionaries
data_total_list = []
data_1_length = len(data_1['news'])
print(data_1_length)
for index in range(0, data_1_length):
    data_dict = {'id': data_1['news'][index]['id'],
                'title': data_1['news'][index]['title'],
                 'url': data_1['news'][index]['url'],
                 'author': data_1['news'][index]['author'],
                 'publication date': data_1['news'][index]['published']
            }
    data_total_list.append(data_dict)

print(data_total_list)
```

```python
data_df = pd.DataFrame(data_total_list)
data_df.to_csv('initialAPIdata.csv', index=False)
data_df
```

# Conducting Sentiment Analysis: NLTK Vader
Reference: https://towardsdatascience.com/sentimental-analysis-using-vader-a3415fef7664

Sources to cite: 
http://comp.social.gatech.edu/papers/icwsm14.vader.hutto.pdf

https://arxiv.org/pdf/1808.01216.pdf

https://arxiv.org/pdf/1704.00939.pdf

```python
data_df['compound score'] = data_df['title'].apply(lambda title: SID.polarity_scores(title)['compound'])
data_df['negative score'] = data_df['title'].apply(lambda title: SID.polarity_scores(title)['neg'])
data_df['positive score'] = data_df['title'].apply(lambda title: SID.polarity_scores(title)['pos'])
data_df['neutral score'] = data_df['title'].apply(lambda title: SID.polarity_scores(title)['neu'])
data_df.head()
```

```python
#Review titles and compound scores to have an initial sense of accuracy of model
for index, row in data_df.iterrows():
    print(row['title'], row['compound score'])
```

```python
#Cursory review of the data suggests that the model is accurate enough to use
#Store in CSV file
data_df.to_csv('initialtestdata.csv', index=False)
```
